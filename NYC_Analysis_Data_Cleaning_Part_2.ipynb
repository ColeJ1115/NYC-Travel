```python
dweather = pd.read_csv('Weather_Cleaned.csv')
```

```python
dweather.head()
```

```python
dweather = dweather.drop(columns= ['EndTime(UTC)'])
```

```python
import datetime as dt
```

```python
dweather['Start_Time'] = pd.to_datetime(dweather['StartTime(UTC)']).dt.date
```

```python
dweather.head()
```

```python
dweather = dweather.drop(columns= ['StartTime(UTC)'])
```

```python
dweather.groupby('Start_Time')['Severity'].count()
```

```python
dweather['Month'] = pd.DatetimeIndex(dweather['Start_Time']).month
```

Filtering weather data most common weather types by month

```python
dweather.groupby('Month')['Type'].max()
```

Total sum of precipitation in inches by month

```python
dweather.groupby('Month')['Precipitation(in)'].sum()
```

```python
dweather.to_csv('Weather_Grouped.csv')
```
