**Merging all Datasets Down to Two Big Ones for Power BI**

```python
import pandas as pd
import numpy as np
```

```python
wea= pd.read_csv('Weather_Grouped.csv')
```

```python
wea.head()
```

```python
wea['Start_Time'] = wea['Start_Time'].astype(str)
```

```python
wea['Period'] = wea['Start_Time'].apply(lambda x: "M" + x[-5:-3])
```

```python
wea.head()
```

```python
wea.tail()
```

```python
wea = wea.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0'])
```

```python
wea.head()
```

```python
wea.head()
```

```python
rfr= pd.read_csv('Hotel_Grouped.csv')
```

```python
rfr.head()
```

```python
rfr = rfr.drop(columns = ['Unnamed: 0', 'Series ID'])
```

```python
rfr.head()
```

```python
wea.to_csv('Weather_Grouped2.csv')
```

```python


weather_df = pd.read_csv('Weather_Grouped2.csv')

merged_df = pd.merge(rfr, weather_df, on='Period')

print(merged_df.head())
```

```python
merged_df = merged_df.drop(columns=['Unnamed: 0'])
```

```python

print(merged_df.columns)
```

```python



print(merged_df.head())
```

```python
merged_df = merged_df.rename(columns={'Start_Time': 'Date'})
```

```python
merged_df.to_csv('Weather.Prices2.csv')
```

```python
nyc= pd.read_csv('LaGuardia_CLEANED_UPDATE2.csv')
```

```python
jfk = pd.read_csv('JFK_CLEANED_UPDATE2.csv')
```

```python
nyc.head()
```

```python
jfk.head()
```

```python
all_df = pd.merge(merged_df, nyc, on='Date')

print(all_df.head())
```

```python
all_df_jfk = pd.merge(merged_df, jfk, on='Date')

print(all_df_jfk.head())
```

```python
all_df.to_csv('LaGuardia.Weather.Prices.csv')
```

```python
all_df_jfk.to_csv('JFK.Weather.Prices.csv')
```

```python

file_path = 'weather_data.xlsx'

wd_df = pd.read_excel(file_path)

print(wd_df)
```

```python
file_path = 'weather_data.xlsx'

wd_df_jfk = pd.read_excel(file_path)

print(wd_df_jfk)
```

```python
wd_df = wd_df.rename(columns={'date': 'Date'})
```

```python
wd_df_jfk = wd_df_jfk.rename(columns={'date': 'Date'})
```

```python

all_df['Date'] = pd.to_datetime(all_df['Date'])


merged_df = pd.merge(all_df, wd_df, on='Date')

print(merged_df.head())
```

```python
all_df_jfk['Date'] = pd.to_datetime(all_df_jfk['Date'])


merged_df_jfk = pd.merge(all_df_jfk, wd_df_jfk, on='Date')

print(merged_df_jfk.head())
```
