**Airline**

```python
da.head()
```

```python
da = da.drop(da[da['Airport Country Code'] != 'US'].index)
```

```python
da.head()
```

```python
da = da.drop(da[(da['Arrival Airport'] != 'LGA') & (da['Arrival Airport'] != 'JFK') & (da['Arrival Airport'] != 'FLU')].index)
```

```python
da.head()
```

```python
da['Departure Date'] = pd.to_datetime(da['Departure Date'].str.replace("-", "/"))
```

```python
from pandas.tseries.offsets import MonthBegin
```

```python
da['Month_Year'] = da['Departure Date'] - MonthBegin(n=1)
```

```python
da.head()
```

```python
da.describe()
```

```python
da.to_csv('Airline_Cleaned.csv')
```
